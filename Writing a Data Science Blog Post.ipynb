I discovered that the 2023 survey does not include gender, therefore, I decided to just look at data from 2017 to 2022. In addition to this, I also decided to include any response that has female or woman in it when looking at the proportions. This was because each year had different values for the genders and I wanted to include anybody that identified as female.

### Clean

I decided to drop all the rows which had null values in as they were not going to be able to be used in my analysis. Plus, as they were not numerical values, I was unable to fill them with the average response. Dropping the missing values may skew my data as I am unable to know what the genders are, so I don't know if they are females (making the calculated proportions less than they should be) or other genders (making the calculated proportions more than they should be).

```python
female_2017=df_2017.dropna(subset=['Gender'], inplace=False)
female_2018=df_2018.dropna(subset=['Gender'], inplace=False)
female_2019=df_2019.dropna(subset=['Gender'], inplace=False)
female_2020=df_2020.dropna(subset=['Gender'], inplace=False)
female_2021=df_2021.dropna(subset=['Gender'], inplace=False)
female_2022=df_2022.dropna(subset=['Gender'], inplace=False)
```

### Analyse

```python
number_females= female_2017.Gender.str.contains(r'Female|Woman').sum()
total_people= len(df_2017)
print(number_females)
print(total_people)
```

As I would be repeating the same piece of code for each year, I decided to create a function which calculated the proprtion of females for each year.

```python
def calculate_proportion_females(df):
    ''' This function finds the proprtion of females in a dataframe
    
        Input:
        df= dataframe containing a column named Gender
        
        Output:
        proportion_females: the proprtion of females in the df
        '''
    number_females= df.Gender.str.contains(r'Female|Woman').sum()
    total_people= len(df)
    proportion_females= round((number_females/total_people)*100, 2)
    
    return proportion_females
```

I then used the function I had created to calculate the proportion of females for each year

```python
prop_2017= calculate_proportion_females(female_2017)
prop_2018= calculate_proportion_females(female_2018)
prop_2019= calculate_proportion_females(female_2019)
prop_2020= calculate_proportion_females(female_2020)
prop_2021= calculate_proportion_females(female_2021)
prop_2022= calculate_proportion_females(female_2022)
```

I created a new dataframe which had the years and the proportions so that I could easily plot the values in a graph.

```python

# assign data of lists.  
data = {'Year': ['2017', '2018', '2019', '2020', '2021', '2022'], 
        'Proportion': [prop_2017, prop_2018, prop_2019, prop_2020, prop_2021, prop_2022]}  
  
# Create DataFrame  
proportion_df = pd.DataFrame(data)  
  
# Print the output.  
print(proportion_df) 
```

### Visualise

I decided to use a line graph to show my findings as it would allow me to easily see the trend of the proprtion of females from 2017 to 2022.

```python
proportion_df.plot(x='Year', y='Proportion', rot = 60)
plt.xlabel("Year", size = 10)
plt.ylabel("Proportion", size = 10)
plt.title("Proportion of female respondants", size = 15)
```

The results show that the proportion of females has decreased quite significantly in 2021 and 2022.

### Clean

I decided to remove the rows of the dataframes that did not have a response, this was because it wouldn't impact my results as there was no way of knowing what these values would have been. As the values are not numerical, it did not make sense to put an average value in. 

```python
employed_2017=df_2017.dropna(subset=['Employment'])
employed_2018=df_2018.dropna(subset=['Employment'])
employed_2019=df_2019.dropna(subset=['Employment'])
employed_2020=df_2020.dropna(subset=['Employment'])
employed_2021=df_2021.dropna(subset=['Employment'])
employed_2022=df_2022.dropna(subset=['Employment'])
employed_2023=df_2023.dropna(subset=['Employment'])
```

### Analyse

I created a function to calculate the proportion of full-time employed.

```python
def calculate_proportion_fullTime(df):
    ''' This function finds the proprtion of full time employed in a dataframe
    
        Input:
        df= dataframe containing a column named Employment
        
        Output:
        proportion_fullTime: the proprtion of full time employed in the df
        '''
    number_fullTime= df.Employment.str.contains(r'Employed, full-time|Employed full-time').sum()
    total_people= len(df)
    proportion_fullTime= round((number_fullTime/total_people)*100, 2)
    
    return proportion_fullTime
```

I used the function to calculate the proportion in each year and then stored them in variables.

```python
full_2017= calculate_proportion_fullTime(employed_2017)
full_2018= calculate_proportion_fullTime(employed_2018)
full_2019= calculate_proportion_fullTime(employed_2019)
full_2020= calculate_proportion_fullTime(employed_2020)
full_2021= calculate_proportion_fullTime(employed_2021)
full_2022= calculate_proportion_fullTime(employed_2022)
full_2023= calculate_proportion_fullTime(employed_2023)
```

Quick check to see if the function has worked.

I then created a dataframe with each of the proportions for each year. This was so it would be easier creating  graph from the data.

```python
# assign data of lists.  
data = {'Year': ['2017', '2018', '2019', '2020', '2021', '2022', '2023'], 
        'Proportion': [full_2017, full_2018, full_2019, full_2020, full_2021, full_2022, full_2023]}  
  
# Create DataFrame  
full_prop_df = pd.DataFrame(data)  
  
# Print the output.  
print(full_prop_df) 
```

### Visualise

I decided a line graph was the best to show the trend of proportions throughout the years.

```python
full_prop_df.plot(x='Year', y='Proportion', rot = 60)
plt.xlabel("Year", size = 10)
plt.ylabel("Proportion", size = 10)
plt.title("Proportion of full-time employed respondants", size = 15)
```

There is a decrease in 2021, then it rises a little in 2022. This graph made me questions whether there was a relationship between females and full-time employed. I therefore, decided to investigate this more.

## Question 3- Is there a relationship between full time employed and proportion of females?

I started with creating a scatter plot of the proportion of females for each year against the proportion of full-time employed. To do this, I had to combine the two dataframes I have made for the female proportion and the full-time employed proportion.

```python
relationship_df=pd.merge(proportion_df, full_prop_df, on='Year')
relationship_df
```

### Visualise

```python
x=relationship_df['Proportion_x']
y=relationship_df['Proportion_y']
plt.title('Relationship between females and full-time employed') 
plt.xlabel('Proportion of females')
plt.ylabel('Proportion of full-time employed')
plt.scatter(x, y)
plt.show()
```

```python
relationship_df.corr()
```

There does appear to be a positive relationship between the proportion of females and the proportion of full-time employed. However, it is not a very strong relationship, with the correlation coefficient of 0.72 (2dp).
